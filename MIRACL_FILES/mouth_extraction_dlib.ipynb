```python
import numpy as np 
import pandas as pd
from imutils import face_utils
import dlib
import cv2
from matplotlib import pyplot as plt
from PIL import Image
```

```python
p = "../input/miraclvc1/shape_predictor_68_face_landmarks.dat"
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(p)
```

```python
inp_dir_path = '../input/miraclvc1/dataset/dataset'
```

```python
def get_rect(shape):
    rw = 0
    rh = 0
    rx = 65535
    ry = 65535
    for (x,y) in shape:
        rw = max(rw,x)
        rh = max(rh,y)
        rx = min(rx,x)
        ry = min(ry,y)
    return (rx,ry,rw-rx,rh-ry)
```

```python
df = pd.DataFrame(columns = ['person_id', 'word_id', 'image_id', 'mouth_path'])
```

```python
df
```

```python
first_words_path = ''
single_word_path = ''
single_word_path_inside = ''
```

```python

for p in [1,2,4,5,6,7,8,9,10,11]:
    if (p==10 or p==11):
        first_words_path = inp_dir_path + '/F' + str(p) + '/words'
    else:
        first_words_path = inp_dir_path + '/F0' + str(p) + '/words'
        
    for i in range(1,11):
        if(i==10):
            single_word_path = first_words_path + '/'+ str(i)
        else:
            single_word_path = first_words_path + '/0' + str(i)


        for j in range(1,11):
            if(j==10):
                single_word_path_inside = single_word_path + '/'+ str(j)
            else:
                single_word_path_inside = single_word_path + '/0' + str(j)

                
            for k in range(1,11):
                try:
                    
                    img_path = single_word_path_inside + '/color_00' + str(k) + '.jpg'
                    
                    image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

                    rects = detector(image, 1)
                    if len(rects) == 0:
                            print("error finding head at file: %s" % img_path)
                            continue

                    shape = predictor(image,box=rects[0])
                    shape = face_utils.shape_to_np(shape)    
                    
                    shape = [shape[x] for x in range(48,68)]
                    (x, y, w, h) = get_rect(shape)
                    
                    mouth = image[y:y+h, x:x+w].copy()
                    mouth = cv2.resize(mouth, (128,128))   
                    out_im = Image.fromarray(mouth)
                    mouth_path = 'F0'+ str(p) + '0' + str(i) + 'mouth'+ str(k) + '.jpg'
                    out_im.save(mouth_path)

                    df = df.append({'person_id':p, 'word_id':j,'image_id':k, 'mouth_path': mouth_path}, ignore_index = True)    

                    
                except:
                    mouth_path = 'F0'+ str(p) + '0' + str(i) + 'mouth'+ str(k) + '.jpg'
                    a = np.zeros((128,128,1),np.uint8)*255
                    a = np.squeeze(a, axis=2)
                    out_im = Image.fromarray(a)
                    out_im.save(mouth_path)                    

                    df = df.append({'person_id':p, 'word_id':j,'image_id':k, 'mouth_path': mouth_path}, ignore_index = True)    
```
